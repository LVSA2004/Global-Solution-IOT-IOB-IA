### Imports

```python
import numpy as np
import pandas as pd
from sklearn.neighbors import KNeighborsClassifier
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import accuracy_score
df = pd.read_csv("FAOSTAT_data_en_6-4-2023.csv")
```

### Colocando o dataframe em ordem pelo Ano, do mais recente ao mais antigo

```python
df = df.sort_values('Ano', ascending=False)
df
```

### Deletando coluna que não utilizaremos

```python
df = df.drop('Código', axis=1)
df = df.drop('Domínio', axis=1)
df = df.drop('M49', axis=1)
df = df.drop('Area', axis=1)
df = df.drop('Source Code', axis=1)
df = df.drop('Year Code', axis=1)
df = df.drop('Source', axis=1)
df = df.drop('Flag', axis=1)
df = df.drop('Flag Description', axis=1)
df = df.drop('Note', axis=1)
df = df.drop('Item Code', axis=1)
df = df.drop('Element Code', axis=1)
df = df.drop('Unit', axis=1)
```

```python
df
```

Transformando os valores das colunas em vetores exemplo = [0,1],[1,0] etc

```python
def one_hot(categorias, valor):
  categorias = list(set(categorias))
  indice = categorias.index(valor)
  vetor = np.zeros(len(categorias))
  vetor[indice] = 1
  return vetor
```

```python
df.Item = [one_hot(df.Item, _) for _ in df.Item]
df.Ano	 = [one_hot(df.Ano	, _) for _ in df.Ano	]
df.Value = [one_hot(df.Value, _) for _ in df.Value]
df.Elemento = [one_hot(df.Elemento, _) for _ in df.Elemento]
```

```python
labels = df.Elemento
df = df.drop(columns = 'Elemento')
```

```python
df
```

```python
def vetorizar(Item, Ano, Value):
    return Item.tolist() + Ano.tolist() + Value.tolist()
```

```python
entrada = [vetorizar(*_) for _ in df.values]
```

### Salvando as tabelas que serão utilizadas no treino em arquivos diferentes

```python
np.savetxt("entrada_treino.csv", entrada, delimiter=',')
```

```python
np.savetxt("labels.csv", labels.tolist(), delimiter=',')
```

```python
x = np.loadtxt("entrada_treino.csv", delimiter=',')
y = np.loadtxt("labels.csv", delimiter=',')
```
